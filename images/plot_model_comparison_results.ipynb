```python
import matplotlib.pyplot as plt

# Define the points
point1 = (4.6, 0.65)
point2 = (4.6, 0.87)
point3 = (6.6, 0.88)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the points with larger dots
plt.scatter(point1[0], point1[1], color=['#00D6C2'], label='Baseline', s=170, marker='s')
plt.scatter(point2[0], point2[1], color=['#7D7D7D'], label='Others', s=80, marker='o')
plt.scatter(point3[0], point3[1], color=['#00D6C2'], label='Others', s=170, marker='s')

# Add labels and title
plt.xlabel('Segmentation Speed [img/sec]', fontsize=14)
plt.ylabel('Dice Coefficient (Similarity)', fontsize=14)
#plt.title('Improvement in Segmentation Performance')

# Add horizontal arrow with label "Better"
plt.annotate('Better', xy=(7.4, 0.98), xytext=(6.5, 0.975), arrowprops=dict(facecolor='black', arrowstyle='->'))

# Add vertical arrow with label "Better"
plt.annotate('Better', xy=(7.4, 0.98), xytext=(7.28, 0.85), arrowprops=dict(facecolor='black', arrowstyle='->'))


# Set x and y limits
plt.xlim(4, 7.5)
plt.ylim(0.6, 1)

# Add legend
plt.legend(['Baseline', 'Other', 'Enhanced'], loc='lower right', fontsize='large', markerscale=1)

plt.savefig("plot_model_comparison_results.png", dpi=600)

# Show plot
#plt.grid(True)
plt.show()
```
